**Feature Engineering**

**As we have 3 columns with categorical value we need to convert the categorical value into the numerical value**

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df["bomb_planted"]=le.fit_transform(df["bomb_planted"]) #These column contain True/False
df["map"]=le.fit_transform(df["map"])  #These contain desert map, city map, terrain map
df["round_winner"]=le.fit_transform(df["round_winner"]) # CT win the round | T win the round
```

```python
df
```

```python
#spliting the data
x=df.iloc[:,:-1] # Featured/Independent Variable
y=df["round_winner"] # Target column/Dependent column
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=34)
```

# **Building model without LDA**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
import warnings
warnings.filterwarnings("ignore") #It will remove all the unwanted warnings

log_model = LogisticRegression()
log_model.fit(x_train, y_train)
log_pred = log_model.predict(x_test)
accuracy_score(y_test, log_pred)
print("Accuracy score of logistic regression model without applying LDA is",accuracy_score(y_test,log_pred)*100)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)  #Scale the independent variables of Training data
x_test=sc.transform(x_test)        #Scale the independent variables of Testing data
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda=LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train,y_train)
```

```python
#when you run lda.transform(x_test), the lda model transsform the test data (x_test) into a new space with reduced dimensions.
lda.transform(x_test)
```

Each feature in our dataset contributes differently to separating the classes. lda.coef_ provides the important for each feature for each class.

These coefficient represent the direction and magnitude by which each feature helps in differentiating between the classes.

```python
lda.coef_
```

```python
lda_coefficients=np.exp(np.abs(lda.coef_)) #Whatever coef value we got, we will find its  absolute then we will find exponential of it.
lda_coefficients #now we got the correct importance of each column out of total 96 columns we were having
```

```python
lda_coefficients=lda_coefficients.flatten() #It convert 2d array to 1d array
lda_coefficients
```

```python
feature_name=x.columns #create a list of column names so we can combine them with column importance
feature_name
```

```python
plt.figure(figsize=(20,10))
plt.bar(feature_name,lda_coefficients) #Create a bar chart of column name vs column improtance
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylabel("Score")
plt.show()
```

```python
df_feature_score=pd.DataFrame({"Features":feature_name,"Score":lda_coefficients})
df_feature_score #DataFrame of column name and its importance
```

```python
top_20_values=df_feature_score.nlargest(20,"Score") #is using the nlargers method to retrieve the top 20 rows from
top_20_values                                       # a DataFrame based on a specific column name "feature_scores"
```
